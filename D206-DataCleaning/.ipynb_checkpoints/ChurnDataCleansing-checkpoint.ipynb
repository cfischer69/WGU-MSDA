```python
#import needed libraries
import pandas as pd
import numpy as np
import missingno as msno
import matplotlib.pyplot as plt

#set some options
#allow pd.describe to show all columns of the data frame
pd.options.display.max_columns = None # https://stackoverflow.com/questions/64455605/show-all-colums-of-a-pandas-dataframe-in-describe 3/8
```

```python
#load church_raw_data.csv
raw_data = pd.read_csv('churn_raw_data.csv')
```

```python
#Examine the data frame
raw_data.describe(include='all') # https://stackoverflow.com/questions/24524104/pandas-describe-is-not-returning-summary-of-all-columns 3/8
# results:
# no duplicates in Customer_id or Interaction
```

```python
raw_data.duplicated().value_counts() # https://note.nkmk.me/en/python-pandas-duplicated-drop-duplicates 3/8
```

```python
#Find features with any missing values
raw_data[raw_data.columns[raw_data.isnull().any()]].isnull().sum() 
# https://stackoverflow.com/questions/36226083/how-to-find-which-columns-contain-any-nan-value-in-pandas-dataframe 3/8
```

```python
#Print the missing data matrix
msno.matrix(raw_data, fontsize = 12, labels=True)
plt.title('Missing Data Matrix')
plt.show()
```

```python
#Sort the data by Age to see if there is a correlation of missingness between Age, Children, Income
msno.matrix(raw_data.sort_values(by='Age'), fontsize = 12, labels=True)
plt.title('Missing Data Matrix by Age')
plt.show()
```
