```python
import matplotlib.pyplot as plt
import pandas as pd
from plotnine import ggplot, aes, geom_bar
from scipy.stats import ttest_ind
import seaborn as sns

churn = pd.read_csv("churn_clean.csv")

churn_yes = churn[churn.Churn == "Yes"].MonthlyCharge
churn_no = churn[churn.Churn == "No"].MonthlyCharge

ttest = ttest_ind(churn_yes,churn_no)

print(ttest)

sns.set()

# Continuous Univariate
fig, axes = plt.subplots(2, 1, figsize=(18, 10))
sns.boxplot(ax=axes[0], x="Outage_sec_perweek", data=churn)
sns.boxplot(ax=axes[1], x="Yearly_equip_failure", data=churn)
plt.show()

# Categorical Univariate
fig, axes = plt.subplots(1, 2, figsize=(18, 10))
sns.countplot(ax=axes[0], x="InternetService", data=churn)
sns.countplot(ax=axes[1], x="Contract", data=churn)
plt.show()

# Continuous Bivariate
sns.jointplot(x="Tenure", y="MonthlyCharge", data=churn)
plt.show()

# Categorical Bivariate
print(ggplot(churn) + aes(x="InternetService", fill="Churn") + geom_bar())
```
